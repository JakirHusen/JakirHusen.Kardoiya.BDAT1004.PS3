```python
# ----------------------------- Question 6 -----------------------------


marriage_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
marriage_data = marriage_data[(marriage_data.Year >= 1867) & (marriage_data.Year <= 2014)]
marriage_data['Marriages_per_capita'] = marriage_data.Marriages / marriage_data.Population
marriage_data['Divorces_per_capita'] = marriage_data.Divorces / marriage_data.Population


plt.plot(marriage_data.Year, marriage_data.Marriages_per_capita, label='Marriages', color='g')
plt.plot(marriage_data.Year, marriage_data.Divorces_per_capita, label='Divorces', color='r')
plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.legend()


plt.show()
```

```python
# ----------------------------- Question 9 -----------------------------

emperor_data = pd.read_csv('roman-emperor-reigns.csv')
total_emperors = len(emperor_data)
assassinated = len(emperor_data[emperor_data['Cause_of_Death'] == 'Assassinated'])
assassinated_percent = (assassinated / total_emperors) * 100
survived_percent = 100 - assassinated_percent


labels = ['Assassinated', 'Survived']
sizes = [assassinated_percent, survived_percent]
colors = ['pink', 'cyan']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Fraction of All Roman Emperors Assassinated')
plt.axis('equal')


plt.show()
```
